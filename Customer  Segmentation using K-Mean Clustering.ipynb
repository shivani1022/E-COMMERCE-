```python
X=df.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
```

```python
for i in range(2,10):
    kmeans= KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(2,10),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. Of Clusters')
plt.ylabel('wcss Values')
plt.show()
```

```python
kmeansmodel=KMeans(n_clusters = 5, init='k-means++', random_state=0)
```

```python
y_kmeans = kmeansmodel.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,1], s= 80, c = "green", label='Customer 1')
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,1], s= 80, c = "yellow", label='Customer 2')
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,1], s= 80, c = "blue", label='Customer 3')
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,1], s= 80, c = "pink", label='Customer 4')
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,1], s= 80, c = "brown", label='Customer 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s= 100, c='black',label='Centroids')
plt.title('Clusters Of Customers')
plt.xlabel('Annual Income (k$)')  
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()            
```
